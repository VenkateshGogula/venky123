```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import silhouette_score
import warnings

# Ignore warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv("Top Indian Places to Visit 2.csv")

# Display the first few rows of the dataset
print(df.head())

# Summary of the dataset
print(df.info())

# 1. Pie Plot function (for 'Type' column)
def create_pie_plot(data, column_name):
    # Count the occurrences of each unique value in the column
    value_counts = data[column_name].value_counts()

    # Create a pie plot
    plt.figure(figsize=(7, 7))
    plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
    plt.title(f'Pie Plot of {column_name}')
    plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
    plt.show()

# Example usage: Pie chart of 'Type' column (categorical feature)
create_pie_plot(df.head(20), 'Type')

# 2. Line Plot function (for 'Establishment Year' vs 'Google review rating')
def create_line_plot(data, x_column, y_column):
    # Convert the 'Establishment Year' to numeric
    data[x_column] = pd.to_numeric(data[x_column], errors='coerce')

    # Create a line plot for continuous data over time
    plt.figure(figsize=(10, 6))
    sns.lineplot(x=data[x_column], y=data[y_column], marker='o', color='b')
    plt.title(f'{y_column} vs {x_column} (Line Plot)')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.grid(True)
    plt.show()

# Example usage: Line plot for 'Establishment Year' vs 'Google review rating'
create_line_plot(df, 'Establishment Year', 'Google review rating')

# 3. Violin Plot function (for 'Google review rating' by 'Significance')
def create_violin_plot(data, x_column, y_column):
    # Create a violin plot to visualize the distribution of a feature by category
    plt.figure(figsize=(8, 6))
    sns.violinplot(x=data[x_column], y=data[y_column], color='lightblue')
    plt.title(f'Violin Plot of {y_column} by {x_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.show()

# Example usage: Violin plot for 'Google review rating' by 'Significance'
create_violin_plot(df, 'Significance', 'Google review rating')
# 4. K-Means Clustering Plot function
def kmeans_clustering_plot(data, features, max_clusters=6):
    # Handle missing values by imputation (mean strategy)
    imputer = SimpleImputer(strategy='mean')
    data[features] = imputer.fit_transform(data[features])
    
    X = data[features]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)  # Standardize the features

    # Determine the optimal number of clusters using the Elbow Method
    wcss = []
    for i in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
        kmeans.fit(X_scaled)
        wcss.append(kmeans.inertia_)

    # Plot the Elbow Method to determine optimal clusters
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, max_clusters + 1), wcss, marker='o', linestyle='--')
    plt.title('Elbow Method for Optimal Clusters')
    plt.xlabel('Number of Clusters')
    plt.ylabel('WCSS')
    plt.show()

    # Apply KMeans with the optimal number of clusters (for simplicity, let's choose 3 clusters)
    optimal_kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
    y_kmeans = optimal_kmeans.fit_predict(X_scaled)

    # Plotting the clusters
    plt.figure(figsize=(8, 6))
    plt.scatter(X_scaled[y_kmeans == 0, 0], X_scaled[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
    plt.scatter(X_scaled[y_kmeans == 1, 0], X_scaled[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
    plt.scatter(X_scaled[y_kmeans == 2, 0], X_scaled[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')

    # Plotting the centroids
    plt.scatter(optimal_kmeans.cluster_centers_[:, 0], optimal_kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids', marker='X')

    # Titles and labels
    plt.title('K-Means Clustering (3 Clusters)')
    plt.xlabel(features[0])
    plt.ylabel(features[1])
    plt.legend()
    plt.show()

# 4. Elbow/Silhouette plot function for clustering (using 'Google review rating' and 'Entrance Fee in INR')
def plot_elbow(data, features, max_clusters=10):
    # Handle missing values by imputation (mean strategy)
    imputer = SimpleImputer(strategy='mean')
    data[features] = imputer.fit_transform(data[features])
    
    X = data[features]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    wcss = []
    for i in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
        kmeans.fit(X_scaled)
        wcss.append(kmeans.inertia_)


    # Elbow method plot
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, max_clusters + 1), wcss, marker='o', linestyle='--')
    plt.title('Elbow Method for Optimal Clusters')
    plt.xlabel('Number of Clusters')
    plt.ylabel('WCSS')
    plt.show()

# Example usage: Elbow plot for clustering on 'Google review rating' and 'Entrance Fee in INR'
plot_elbow(df, ['Google review rating', 'Entrance Fee in INR'])

# 5. Linear Regression Fit function (for predicting 'Google review rating' based on 'Entrance Fee in INR')
def linear_regression_fit(data, x_column, y_column):
    # Handle missing values for features
    imputer = SimpleImputer(strategy='mean')
    X = data[[x_column]].values
    y = data[y_column].values
    
    # Impute missing values for features
    X_imputed = imputer.fit_transform(X)

    # Initialize the linear regression model
    model = LinearRegression()
    model.fit(X_imputed, y)

    # Predictions
    y_pred = model.predict(X_imputed)

    # Plotting the regression line
    plt.figure(figsize=(8, 6))
    plt.scatter(X_imputed, y, color='green', label='Actual')
    plt.plot(X_imputed, y_pred, color='red', label='Regression Line')
    plt.title(f'Linear Regression: {y_column} vs {x_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.legend()
    plt.show()

    # Displaying model metrics
    print(f'Coefficients: {model.coef_}')
    print(f'Intercept: {model.intercept_}')
    print(f'Mean Squared Error: {mean_squared_error(y, y_pred)}')
    print(f'R² Score: {r2_score(y, y_pred)}')

# Example usage: Linear Regression on 'Entrance Fee in INR' vs 'Google review rating'
linear_regression_fit(df, 'Entrance Fee in INR', 'Google review rating')
# 4. K-Means Clustering on 'Google review rating' and 'Entrance Fee in INR'
kmeans_clustering_plot(df, ['Google review rating', 'Entrance Fee in INR'])
```
